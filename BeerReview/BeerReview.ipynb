## 결측치 제거 및 데이터 정리

- brewery_name 컬럼: 양조장 이름은 대신할 수 있는 컬럼(brewery_id)가 있으므로, 컬럼 삭제 처리.
- review_profilename 컬럼: 리뷰를 작성한 사람의 이름은 대신할 수 있는 컬럼이 없으므로, 결측치 삭제 처리.
- beer_abv 컬럼: 알콜 도수 데이터를 구할 수 없으므로, 결측치 삭제 처리.
- beer_name 컬럼: 맥주 이름은 대신할 수 있는 컬럼(beer_beerid)가 있으므로, 컬럼 삭제 처리.

```python
df = origin.drop(columns=['brewery_name']).copy()
print(df.shape)
df.head(10)
```

```python
df = df[~df['review_profilename'].isnull()]
print(df.shape)
```

```python
df = df[~df['beer_abv'].isnull()]
print(df.shape)
```

```python
df = df.drop(columns=['beer_name'])
print(df.shape)
df.head(10)
```

## 데이터 유형 탐색

```python
df.info()
```

- brewery_id, beer_beerid 컬럼: category 형으로 변환.

```python
df['beer_beerid'] = df['beer_beerid'].astype('category')
```

## 파생 변수 생성

### 총점 변수 생성

- 총 점은 각각의 리뷰 점수(외관, 맛, 향, 느낌, 전반적인 점수)의 평균으로 메김.

```python
review_scores = ['review_overall', 'review_aroma', 'review_appearance', 'review_palate', 'review_taste']
```

```python
df['review_score'] = df[review_scores].sum(axis=1) / 5
print(df.shape)
df.head(10)
```

```python
review_scores.append('review_score')
```

## 상관계수 확인하기

```python
df[review_scores[:-1] + ['beer_abv', 'beer_abv_log1p']].corr()
```

- review_palate(느낌으로 추측), review_taste(맛) 컬럼의 값이 높을 수록 review_overall(전반적인 점수)를 높게 평가하였다.
- review_taste(맛)와 review_aroma(향)은 서로 높은 상관관계를 가지고 있었다.
- review_taste(맛)와 review_palete(느낌)은 서로 높은 상관관계를 가지고 있었다.

- beer_abv_log1p(알콜 도수)와 review_appearance(외관)은 유의미하게 관계를 가지고 있는 것 같지는 않다.

- 대체로 review_taste(맛)의 점수가 높을 수록 다른 점수도 높을 것으로 추측해볼 수 있다.

## 양조장 별 분석

### 양조장 별 평균 평점

```python
df_brew_scores = df.groupby(['brewery_id']).agg({_: 'mean' for _ in review_scores}).reset_index()
print(df_brew_scores.shape)
df_brew_scores.head(10)
```

```python
df_brew_scores.nlargest(5, 'review_score')
```

아무래도, 1~2개만 생산한 양조장의 점수가 제일 높게 나온 것 같아, 리뷰의 개수를 포함하여 순위를 매기려고 한다.

```python
review_scores_dict = {_: 'mean' for _ in review_scores}
review_scores_dict['review_time'] = 'count'
review_scores_dict
```

```python
df_brew_scores = df.groupby(['brewery_id']).agg(review_scores_dict).reset_index()
print(df_brew_scores.shape)
df_brew_scores.head(10)
```

```python
df_brew_scores['review_time_log1p'] = np.log1p(df_brew_scores['review_time'])
```

```python
np.log1p(df_brew_scores['review_time']).hist(bins=50)
```

```python
df_brew_scores.nlargest(5, 'review_score')
```

```python
df_brew_scores['review_score_3'] = df_brew_scores['review_score'] * df_brew_scores['review_time']
```

```python
df_brew_scores.nlargest(5, 'review_score')
```

```python
df_brew_scores.nlargest(5, 'review_score_2')
```

```python
df_brew_scores.nlargest(5, 'review_score_3')
```

```python
df_brew_scores.nlargest(5, 'review_time')
```
