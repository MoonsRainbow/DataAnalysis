## 양조장 별 분석

### 양조장 별 평균 평점

```python
df_brew_scores = df.groupby(['brewery_id']).agg({_: 'mean' for _ in review_scores}).reset_index()
print(df_brew_scores.shape)
df_brew_scores.head(10)
```

```python
df_brew_scores.nlargest(5, 'review_score')
```

- 리뷰가 1~2개만 있는 양조장이 유리하게 나왔다. 
- 리뷰의 개수를 포함하여 순위를 다시 매겨보자.

### 양조장 별 평균 평점 (feat. 리뷰 개수 반영)

- 순위를 매기는 기준을 5가지로 나누어보려고 한다.<br><br>
    1. 평균 평점
        - 리뷰의 점수가 반영되지 않아, 리뷰가 1개이면서 5점을 기록한 양조장이 가장 순위가 높았다.<br><br>
    1. 평균 평점 * 리뷰 개수
        - 리뷰가 가장 많은 양조장이 유리할 것으로 보인다.<br><br>
    1. 평균 평점 * 리뷰 개수의 log1p
        - 리뷰 개수의 차이를 좁혀보면 어떨지 시도해보려고 한다.<br><br>
    1. 평균 평점 * 리뷰 개수의 log1p 2번
        - 이 부분은 추후에 추가되었다. log1p를 1번 적용했을 때, 역시나 리뷰의 개수에 많은 영향을 받았다.
        - 리뷰 개수의 영향을 좀 더 약화시키기 위해 2번 적용해보려고 한다.<br><br>
    1. 평균 평점 * 리뷰 개수의 cut 구간 별 가중치
        - 리뷰가 너무 많은 양조장이 유리하므로, 리뷰가 적을 수록 높은 가중치를 적용해보려고 한다.

```python
review_scores_dict = {_: 'mean' for _ in review_scores}
review_scores_dict['review_time'] = 'count'
review_scores_dict
```

```python
df_brew_scores = df.groupby(['brewery_id']).agg(review_scores_dict).reset_index()
print(df_brew_scores.shape)
df_brew_scores.head(10)
```

```python
df_brew_scores['rt_log1p'] = np.log1p(df_brew_scores['review_time'])
```

```python
df_brew_scores['rt_log1p_double'] = np.log1p(df_brew_scores['rt_log1p'])
```

- 리뷰의 개수를 반영하지 않으면, 리뷰의 개수가 적으면서 높은 평점을 받은 양조장이 유리하다.
- 리뷰의 개수를 반영하면, 리뷰의 개수가 많을 수록 유리하다.
<br><br>
따라서, 리뷰의 개수를 반영하되, 있는 그대로의 리뷰 개수가 아닌, <br>
어느정도 스케일링이 되거나, 일정 구간에 어느 정도의 어드밴티지를 주어야할 것으로 보인다.
<br><br>
구간 별 Max 중에서 가장 낮은 값(평점)이었던, 4.37 점을 총점으로 만들어 평가해보려고 한다.

- 구간 1. (0 ~ 2개의 리뷰)
    - 이 구간은 리뷰 개수가 많지 않기 때문에, 쉽게 평점 5.0을 받을 수 있다.
    - 따라서, 평점의 총 점을 4.37이 아닌 4.35로 맞추어주었다.<br><br>
    
- 구간 2. (3 ~ 12개의 리뷰)
    - 이 구간 역시 리뷰의 개수가 많지 않기 때문에, 쉽게 높은 평점을 받을 수 있다.
    - 하지만, 5.0점인 양조장은 없다. 따라서 총점을 4.37점에 맞추어주었다.<br><br>
    
- 구간 3. (13 ~ 69개의 리뷰), 구간 4. (70 ~ 1553개의 리뷰), 구간 5. (1554 ~ 38806개의 리뷰)
    - 이 구간부터는 높은 평점을 보기가 쉽지 않다.
    - 하지만, 5.0점인 양조장은 없다. 따라서 총점을 4.37점에 맞추어주었다.<br><br>

```python
bins = [0, 3, 13, 70, 1554, 38806]
labels = [0.87, 0.95, 1, 0.979, 0.971]
```

```python
df_brew_scores['qcut_score'] = pd.cut(df_brew_scores['review_time'], bins, labels=labels)
df_brew_scores['qcut_score'] = df_brew_scores['qcut_score'].astype(float)
print(df_brew_scores.shape)
df_brew_scores.tail(10)
```

#### 평균 평점 기준 랭크

```python
df_brew_scores['score_1'] = df_brew_scores['review_score']
df_brew_scores.nlargest(10, 'score_1')
```

#### 평균 평점 * 리뷰 개수 기준 랭크

```python
df_brew_scores['score_2'] = df_brew_scores['review_score'] * df_brew_scores['review_time']
df_brew_scores.nlargest(10, 'score_2')
```

#### 평균 평점 * 리뷰 개수 log1p

```python
df_brew_scores['score_3'] = df_brew_scores['review_score'] * df_brew_scores['rt_log1p']
df_brew_scores.nlargest(10, 'score_3')
```

#### 평균 평점 * 리뷰 개수 log1p 2번

```python
df_brew_scores['score_4'] = df_brew_scores['review_score'] * df_brew_scores['rt_log1p_double']
df_brew_scores.nlargest(10, 'score_4')
```

#### 평균 평점 * 리뷰 개수의 cut 구간 별 가중치

```python
df_brew_scores['score_5'] = df_brew_scores['review_score'] * df_brew_scores['qcut_score']
df_brew_scores.nlargest(10, 'score_5')
```
